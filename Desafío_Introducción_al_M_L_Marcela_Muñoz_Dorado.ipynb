# **PREGUNTA Nº 1: ¿CUÁL ES LA DIFERENCIA ENTRE EL APRENDIZAJE SUPERVISADO Y EL APRENDIZAJE NO SUPERVISADO?**

# **CONCEPTOS**
**Aprendizaje Supervisado**: Método, técnica o enfoque utilizado por el  Machine Learning (siendo esta una subrama de lA) compuesto por conjuntos de datos de entrada y de salida etiquetados, es decir, cada muestra de estos contiene la respuesta o resultaado esperado o predefinido. Estos conjuntos de datos entrenan algoritmos que tienen por objeto clasificar datos o predecir resultados de manera precisa. El objetivo de este tipo de aprendizaje es entrenar el modelo creado con los datos de entrenamiento ingresados a fin de que aprenda a reconocer caracísterticas, patrones y/o relaciones al momentos de recibir nuevos datos de entrada no etiquetados, y así entregar la salida respectiva o esperada. En otras palabras, se entrena al modelo para que aprenda a mapear las características, patrones y/o relaciones de los datos ingresados a las salidas predefinidas (debe aprender a mapear las entradas a las salidas, basándose en los ejemplos de entrenamientos que se le proporcionaron previamente). Se le llama supervisado porque requiere de la intervención humana principalmente en la etapa de preprocesamiento de los datos de entrenamiento y en su etiqueado, como también para el entrenamiento y ajuste del modelo. A su vez, el apredizaje supervisado se puede agrupar en dos tipos: clasificación y regresión.


**Aprendizaje No Supervisado**: "Método, técnica o enfoque utilizado por el  Machine Learning (siendo esta una subrama de lA) compuesto por conjuntos de datos o etiquetados, estos es, datos que no contienen respuestas o resultados predifinidos o deseados. El algoritmo creado tiene por objeto examinar y/o agrupar a estos conjuntos de datos por medio del reconocimiento de patrones o estructuras en ellos. En el aprendizaje no supervisado existen tres principales categorás de algorimos: Agrupaciones de clústeres, Asociación y Reducción de dimensionalidad.

# **DIFERENCIAS:**

1.   **Forma en que se les proporciona la información**: La forma en que se les entrega información a los algoritmos durante la fase de entrenamiento varía en cada tipo de aprendizaje. En el supervisado se entregan conjuntos de datos con entradas y salidas que contiene una etiqueta o resultado esperado y predefinido, es decir, datos etiquetados. En el caso del aprendizaje no supervisado, los datos proporcionados no se encuentran etiquetados, por ende, no contiene un resultado específico definido.
2.   **Objetivo**: El objetivo del aprendizaje supervisado es aprender una función que mapee las entradas a las salidas, y para ello se basa en los datos y ejemplos de entrenamiento que se le proporcionaron en la etapa de su creación, entrenamiento y ajuste. El objetivo del aprendizaje no supervisado es el reconocimiento de patrones y/o estructuras intrínsecas o subyacentes en los datos que se le otorgan, es decir, debe aprender a descubrir patrones desconocidos en los datos ingresados.
3.   **Intervención Humana**: En los modelos de aprendizaje supervisado se requiere mayor intervención humana en la etapa de procesamiento de datos ya que se debe garantizar que las etiquetas asignadas a los mismos cotengan información adecuada, y ello porque un modelo supervisado dependen, en gran medida, de la capacidad que tenga de conocer el resultado y así concluir si los datos entregados posteriormente infieren o no en él. Los modelos de aprendizaje no supervisados operan si intervención humana constante pues funcionan econtrando estructuras o patrones llegando a ellas mediante datos no etiquetados. Se podría decir que la única ayuda humana se da en la etapa de validación de las variables de los resultados.
4.   **Clasifición**: Los modelos de aprendizaje supervisado se pueden agrupar en dos tipos fundamentales: Clasificación (algoritmos que clasifican datos en segmentos particulares) y de Regresión (Método estadístico y de ML que utiliza algoritmos para medir la relación entre una variable dependiente y una o más variables independientes). Los algoritmos de aprendizaje no supervisado se clasifican en dos grandes categorías: Clustering (agrupación de datos) y Reducción de dimensionalidad (reducción de características o variables de un conjunto de datos).
5.   **Utilización**: En los modelos de aprendizaje supervisado se busca predecir una etiqueta de clase o predecir un valor numérico de los nuevos datos ingresados (clasificación y regresión). En los modelos de aprendizaje no supervisado se busca la comprensión de la estructura o de los patrones de los datos (para segmentar o clasterizar).
5.   **Resultados**: En los modelos de aprendizaje supervisado el resultado esperado es la predicción de una etiqueta de clase (algoritmo de clasificación) o predicción de un valor numérico (algoritmo de regresión), es decir, clasifica o predice un resultado predefinido. De igual forma, se conoce el qué resultados se pueden esperar. En los modelos de aprendizaje no supervisado, el resultado es el descubrimiento de patrones intrínsecos o de estructuras subyacentes de los datos que se le ingresen. Aquí, se desconoce el resultado, pues se descubre algo nuevo.y desconocido.
6.   **Precisión**: Los modelos de aprendizaje supervisado son más precisos toda vez que contienen una salida predefinida y entrenada. En el caso de los algoritmos de aprendizaje no supervisado no existe una precisión alta dado que el resultado o salida es desconocido.

# **PREGUNTA Nº3: ¿POR QUÉ RAZÓN ES NECESARIO DIVIDIR EL SET DE DATOS EN DOS SET DE DATOS AL MOMENTO DE REALIZAR ALGÚN MODELAMIENTO SUPERVISADO? ¿CUÁLES SON LOS DOS SET DE DATOS Y CUÁL ES LA FUNCIÓN DE CADA UNO?**

# **RESPUESTA**
---
####Los nombres que reciben los datos al ser divididos en dos partes (70%-30% u 80%-20% generalmente) son:

1.   **Conjunto de entrenamiento, Training set o training data**: Son el porcentajo de los datos que utilizamos para entrenar un modelo. La calidad de nuestro modelo de aprendizaje automático va a ser directamente proporcional a la calidad de los datos, y he aquí la importancia de la etapa de preprocesamiento de los datos.
2.   **Conjunto de prueba, Validación, Test set o Testing Data**: Se tata de aquella fracción de los datos que se reservan para comprobar si el modelo generado a partir de los datos de entrenamiento funciona. Es decir, si las respuestas predichas por el modelo para un caso totalmente nuevo son acertadas o no. Es de vital importancia que el conjunto de datos de prueba tenga un volumen suficiente como para generar resultados estadísticamente significativos, y a la vez, que sea representativo del conjunto de datos global.

####La razón principal para proceder a su división obedece a motivos de calidad, pues con esta práctica se logra evaluar tanto el rendimiento del modelo como la capacidad que presenta para generalizar nuevos datos que se le otorguen posteriormente.
####Debemos tener en consideración que los algoritmos utilizados en la creación de un modelo aprenderán a reconocer o mapear con los datos que se les ingresen en la fase de entrenamiento, por lo que si ingresáramos la totalidad de los datos con fines de entrenamiento y posteriormente lo probamos o testeamos con los mismos datos, el modelo simplemente memorizaría los datos de entrenamiento y resultaría incapaz de aprender a reconocer otros patrones o estructuras por sí mismo o bien al correr prsentaría una muy baja capacidad para hacer predicciones precisas sobre datos nuevos y no vistos.
####Generalmente, la mayor parte de los datos se utilizan para entrenar el modelo. Suelen repartirse en un 70% de datos de entrenamiento y un 30% de datos de test, pero se puede variar la proporción según el caso. Lo importante es evitar el sobreajuste u “overfitting”, e decir, sobre-entrenar al modelo de manera que se ajusten en demasía al conjunto de entrenamiento, dado que pueden perder gran parte de su capacidad predictiva.
####De ahí surge la importancia de contar con un conjunto de prueba independiente, y de esta forma verificar si el modelo se encuentra sobre-ajustado o no.
####También resulta relevante dividir los datos en ambos conjuntos para utilizar técnicas de validación cruzada y así obtener estimaciones más robustas respecto al rendimiento del modelo y así ajustar los hiper parámetros del modelo en forma más precisa.



# **PREGUNTA Nº4: ¿CUÁL ES LA PRINCIPAL DIFERENCIA EN EL MODELAMIENTO DE LA REGRESIÓN LINEAL PARA UN ENFOQUE ESTADÍSTICO Y UN ENFOQUE DE APRENDIZAJE DE MÁQUINAS?**.

# **RESPUESTA**


---
####Determinar lo anterior dependerá, en gran medida, del objeto que se tenga en vista al momento de crear el modelo de regresión lineal.

####**Enfoque estadístico**: Un enfoque estadístico tiene por objeto la interpretación de las relaciones entre las diversas variables de entrada (predictoras) y la variable de salida o respuesta. Lo anterior se lleva a cabo por medio de un análisis detallado de los parámetros del modelo y de sus propiedades estadísticas. Bajo este prisma, se presta mayor atención a la inferencia estadística, esto es, a la interpretación de los coeficientes de un modelo, sus intervalos de confianza y al p-values a objeto de comprender la significamcia de las relaciones entre las variables. También se ha de considerar que el objeto principal de este enfoque es, generalmente, intentar explicar la relación entre las variables y comprender la naturaleza de las asociaciones entre ellas, como también hacer predicciones.

####**Enfoque de Machine Learning**: Se centra mayoritariamente en la capacidad predictiva que tiene un modelo, esto es, en la capacidad de realizar predicciones precisas sobre datos no vistos (en los conjuntos de entrenamiento y prueba). Este enfoque presta menor atenció a la interpretación en detalle de los parámetros del modelo y se centra más en la optimizaación del rendimiento del modelo según una métrica específica (EJ: MSE, MAE, RMAE, entre otras). Aquí las suposiciones estadísticas pueden ser consideradas en menor medida e incluso excluidas. En este orden de ideas, en el enfoque ML se busca crear un modelo que minimice al máximo el error de predicción al momento de ingresar al mismo nuevos datos.

####Podrámos concluir que el enfoque estadístico de la regresión lineal se centra más en la interpretación y en la inferencia estadística, mientras que en enfoque del ML se centra en la capacidad de predicción y en la optimización del rendimiento de un modelo al momento de ingresar u otorgar nuevos datos.

#CUARTO PASO: REEMPLAZO DE LOS VALORES ATÍPICOS POR LA MEDIA DE CADA VARIABLE.

```python
import pandas as pd
import numpy as np

def eliminar_outliers(data, columnas):
    """
    Elimina los valores outliers en las columnas especificadas de un DataFrame.
    Args:
        data (pandas.DataFrame): DataFrame que contiene los datos.
        columnas (list): Lista de nombres de columnas en las que se eliminarán los outliers.
    Returns:
        pandas.DataFrame: DataFrame con los outliers reemplazados por la media en las columnas especificadas.
    """
    # Trabajar con una copia del DataFrame para evitar cambios en los datos originales
    data = data.copy()

    for columna in columnas:
        # Verificar si la columna contiene datos numéricos
        if pd.api.types.is_numeric_dtype(data[columna]):
            # Calcular cuartiles y rango intercuartílico
            p_25 = data[columna].quantile(0.25)
            p_75 = data[columna].quantile(0.75)
            IQR = p_75 - p_25

            # Calcular límites
            lim_inf = p_25 - 1.5 * IQR
            lim_sup = p_75 + 1.5 * IQR

            # Reemplazar outliers por la media
            data[columna] = np.where((data[columna] < lim_inf) | (data[columna] > lim_sup),
                                     data[columna].mean(), data[columna])
        else:
            print(f'La columna "{columna}" no contiene datos numéricos y no se procesará.')

    return data

# Uso
columnas_a_limpiar = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'guestroom', 'hotwaterheating', 'airconditioning', 'parking']
df_housing_cleaned = eliminar_outliers(df_housing, columnas_a_limpiar)
```

```python
#DIVISIÓN DE LOS DATOS, NORMALIZACIÓN, ENTRENAMIENTO Y CÁLCULO DE LAS MÉTRICAS
#DEL MODELO DE REGRESIÓN LINEAL CREADO PREVIAMENTE.

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error


x = df_housing[['area', 'bedrooms', 'bathrooms', 'stories', 'parking']]


#División de datos en conjunto de entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#Normalización de datos
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

#Entrenamiento del modelo en los datos de entrenamiento
modelo = LinearRegression()
modelo.fit(x_train_scaled, y_train)

#Predicción en los datos de entrenamiento y prueba
y_train_predic = modelo.predict(x_train_scaled)
y_test_predic = modelo.predict(x_test_scaled)

#Métricas de evaluación en los datos de entrenamiento
mse_train = mean_squared_error(y_train, y_train_predic)
mae_train = mean_absolute_error(y_train, y_train_predic)
r2_train = r2_score(y_train, y_train_predic)

#Métricas de evaluación en los datos de prueba
mse_test = mean_squared_error(y_test, y_test_predic)
mae_test = mean_absolute_error(y_test, y_test_predic)
r2_test = r2_score(y_test, y_test_predic)

#Impresión de métricas
print('Métricas en conjunto de entrenamiento:')
print('MSE:', mse_train)
print('MAE:', mae_train)
print('R^2:', r2_train)
print('\nMétricas en conjunto de prueba:')
print('MSE:', mse_test)
print('MAE:', mae_test)
print('R^2:', r2_test)

# Interpretación de coeficientes
print('\nCoeficientes del modelo:')
for feature, coef in zip(x.columns, modelo.coef_):
    print(feature, ':', coef)
print('Intercept:', modelo.intercept_)
```

```python
#CREACIÓN DE FUNCIÓN QUE EVALUA EL MODELO CREADO PARA DF_HOUSINGY GRAFICA SUS COEFICIENTES.

#Variable respuesta y variables predictoras.
X = df_housing[['area', 'bedrooms', 'bathrooms', 'stories', 'parking']]
y = df_housing['price']

#División (80% train, 20% test).
X_train, X_test, y_train, y_test = train_test_split(df_housing[['area', 'bedrooms', 'bathrooms', 'stories', 'parking']], df_housing['price'], test_size=0.2, random_state=42)

#Agregación intercepto para los conjuntos de entrenamiento y prueba.
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)

#Creación y ajuste del modelo con los datos de entrenamiento.
modelo_regresion = sm.OLS(y_train, X_train)
resultados = modelo_regresion.fit()


def evaluacion_del_modelo(model, X_test, y_test):
    #Utilización datos de prueba para realizar predicciones
    y_pred = model.predict(X_test)

    #Evaluación del rendimiento del modelo comparando las predicciones con los valores reales.
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    #Cálculo coeficientes.
    coeficientes = model.params

    #Graficación coeficientes.
    plt.figure(figsize=(10, 6))
    plt.bar(coeficientes.index, coeficientes.values)
    plt.title('Coeficientes del Modelo')
    plt.xlabel('Variables')
    plt.ylabel('Coeficientes')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

    return mse, r2

mse_test, r2_test = evaluacion_del_modelo(resultados, X_test, y_test)
print(f'Error Cuadrático Medio (MSE) - Prueba: {mse_test:.2f}')
print(f'Coeficiente de Determinación (R²) - Prueba: {r2_test:.2f}')
```
